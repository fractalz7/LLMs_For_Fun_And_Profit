```python
import whisper
model = whisper.load_model("base")
result = model.transcribe("sample_audio.mp3")
print(f' The text in video: \n {result["text"]}')
```

StableDiffusion
https://stability.ai/

```python
pipeline = DiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
```

```python
prompt= "can you please sketch an elephant family strolling around Times Square?"
```

```python
image=pipeline(prompt, num_inference_steps=5)
```
